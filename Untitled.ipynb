```python
import pandas as pd

audi= pd.read_csv("./bil_torvet/audi_links.csv",sep=";", header=None)
audi=audi.drop([5],axis=1)
audi
```

```python
for index in audi.index:
    audi.iloc[index,2]=audi.iloc[index,2].replace('.','')
    if(audi.iloc[index,2].isdigit()):
        audi.iloc[index,2]=float(audi.iloc[index,2])
    else:
        audi.iloc[index,2]=None
        
    audi.iloc[index,0]=audi.iloc[index,0].replace('Audi','').strip()

bool_matrix=audi.isnull() 
only_null_filter = bool_matrix.any(axis=1)
missing = audi[only_null_filter]
audi.dropna(inplace = True)
audi
```

```python
audi=pd.get_dummies(audi,columns=[0,3])
audi
```

```python
list(audi.columns)
```

```python
from sklearn.cluster import estimate_bandwidth
bw=estimate_bandwidth(audi)
bw
```

```python
from sklearn.cluster import MeanShift
import numpy as np

analyzer= MeanShift(bandwidth=bw)
analyzer.fit(audi)
labels= analyzer.labels_
centers=analyzer.cluster_centers_
print(np.unique(labels))
print(centers)
```
